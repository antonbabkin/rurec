<h2><center>Location and Rurality Measures in InfoGroup Extracts</center></h2>

<center>Included in InfoGroup</center>

##### Address Line 1
The street address, name and number.
    
##### City
The name of the city as a text string.
    
##### County Code
The 3-digit FIPS code identifying a county uniquely within a state.
    
##### FIPS Code
The 5-digit FIPS code identifying a county uniquely within the U.S.
    
##### State
The 2-character postal state name abbreviation. InfoGroup does not include the 2-digit state 
FIPS code.
    
##### ZipCode
The 5-digit postal zip code. <span style="color:#fc3d03">Postal Zip codes can cross county 
and state boundaries.</span>
    
##### Zip4
The 4-digit add-on to the postal zip code.
    
##### Area Code
The 3-digit telephone area code. <span style="color:#fc3d03">Area codes do not cross state 
boundaries.</span>
    
##### Census Tract
##### Census Block
<span style="color:#fc3d03">Tracts are contained within a county, blocks within a tract.
Tracts and blocks do not necessarily respect other political boundaries.</span> 
'Census Tract' is a 6-digit numeric string identifying the tract as unique within a county. 
'Census Block' is a 3-digit numeric string that is unique within a tract.

Census tracts and blocks are subject to revision every ten years.

##### Full Census Tract
This variable is created out of the state, county, and census tract identifiers included
in the InfoGroup file, resulting in an 11-digit numeric string that identifies a census
tract uniquely within the country.

##### Latitude
##### Longitude
Imputed presumably from a sufficient combination of the locational variables above.

<center> OMB Metropolitan/Micropolitan Concepts</center>

##### CBSA Code
CBSA refers to the OMB’s Core-Based Statistical Area, which comprises two categories of 
“urban” area: metropolitan and micropolitan. 
    
<span style="color:#fc3d03">A CBSA consists of one or more counties</span> that are 
“anchored” by an urban center of at least 50,000 people (“metropolitan”) or 10,000-50,000 
people (“micropolitan”), “plus adjacent counties that are socioeconomically tied to the 
urban center by commuting”. <span style="color:#fc3d03">The counties that comprise a CBSA 
do not have to be in the same state.</span>

CBSA Code is the OMB's unique 5-digit identifier. An InfoGroup
record that is not CBSA-based is coded '00000'.

Refer for further information about CBSA to:                             
https://www.census.gov/topics/housing/housing-patterns/about/core-based-statistical-areas.html

##### CBSA Level
CBSA Level identifies a CBSA as metropolitan ("2") or micropolitan ("1"). 
Missing on this variable conventionally denotes a 'rural' observation.

##### CSA Code
A CSA, or Combined Statistical Area, is <span style="color:#fc3d03">a combination of 
CBSAs</span> (either metro or micro)
that are demonstrably linked by some degree of economic exchange, including 
commuting patterns. The linkage is by definition less than that which defines the metropolitan
and micropolitan areas themselves. 

The CSA code is a 3-digit number string. An InfoGroup record for an establishment that is not
located in a CSA is coded '000'.

For further information and references to technical material about CSA, start with:
https://en.wikipedia.org/wiki/Combined_statistical_area

<center>Created Using External Data Sources</center>

<center>U.S. Census Urban Area Concepts</center>

##### UA Code
Created by matching the Urban Area ("UA") code from this 2010 Census relationship file        
http://www2.census.gov/geo/docs/maps-data/data/ua-cbsa_rel_10.txt             
to the InfoGroup CBSA Code value on each record. 

The relationship file is released by the Census Bureau, and Urban Areas are subject to 
redefinition, in each federal census year. 

<p style="color:#fc3d03">An Urban Area is a collection of census tracts (sometimes census 
blocks). A UA is not congruent with any other statistical or administrative spatial unit.</p>

For further information about the Urban Area concept, refer to:                
https://www.census.gov/programs-surveys/geography/guidance/geo-areas/urban-rural/2010-urban-rural.html

##### UA Type
An Urban Area can belong to one of two types: Urbanized Area or Urban Cluster. The details
of definition of these types are different from those of the metropolitan/micropolitan duality
among CBSAs, but the theoretical premise is very similar. An Urbanized Area is defined by a more 
densely settled spatial core than in an Urban Cluster. Urban Areas are on average considerably
smaller than CBSAs.

The UA-to-CBSA relationship file contains a text description that identifies each UA as either
an Urbanized Area or an Urban Cluster.

##### rural_by_UA
This is a wholly created flag that identifies whether the Urban Area attributed to an
InfoGroup record, being smaller than a county or CBSA, overlaps (or is contained in) rural 
or urban counties or both. An urban county is one that is part of a CBSA.

<center>ERS Far and Remote Concepts</center>

##### far1, far2, far3, far4

##### FAR Level

The USDA writes: “To assist in providing policy-relevant information about conditions in 
sparsely-settled, remote areas of the U.S. to public officials, researchers, and the general 
public, ERS has developed <span style="color:#fc3d03">ZIP-code-level</span> frontier and 
remote area (FAR) codes”.

FAR codes are applied to postal zip codes to identify different degrees and criteria of 
remoteness. It is not a code for any functional concept of rurality, but there is an obvious 
family resemblance between “remote” and “rural” which might find some analytical use.

The ERS created four FAR levels based on proximity (conceived of as travel time) to 
“urban” places of different sizes. Levels 1 through 4 measure increasing remoteness. The 
‘far1’ through ‘far4’ variables are 1/0 flags indicating whether the InfoGroup’s zip code is 
classified in at least that degree of remoteness. The ‘FAR Level’ variable captures the highest 
numbered positive FAR flag for a location. For example, a record with 1’s in ‘far1’, ‘far2’, 
and ‘far3’, but a 0 in ‘far4’, would have a ‘FAR Level’ value of 3.

A detailed description of FAR level criteria is on the ERS web page 

https://www.ers.usda.gov/data-products/frontier-and-remote-area-codes/documentation/.



<center>Rurality of Census Tracts</center>

##### rural_HRSA

The HRSA (Health Resources and Services Administration), in particular its sub-unit, the 
Federal Office of Rural Health Policy (FORHP), refines the OMB definition of rural by 
adding rural census tracts otherwise located in metro/micro 
CBSA counties to the tracts located in non-CBSA counties. It does not distinguish between 
rural and urban census tracts in non-CBSA counties.

The FORHP uses the ERS’s Rural-Urban Commuting Area (RUCA) criteria for defining 
rurality in CBSA census tracts.

For more information about the HRSA/FORHP rurality concept, go to               
https://data.hrsa.gov/data/about                                          
and scroll down to 'FORHP Rural Areas'.

This is a 1/0 flag variable. 1 = rural, 0 = urban.

##### rural_tract

The criterion of rurality in this variable is whether the spatial centroid of the census 
tract falls within the boundaries of a Census Urban Area (UA).

This is a laboriously computed 1/0 flag variable coded 1 = rural, 0 = urban.

A detailed description of the method applied to create this variable is
included in the 'Other Spatial Concepts' notebook.
